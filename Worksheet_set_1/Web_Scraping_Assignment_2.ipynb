```python
# Q1) Python program to scrape data for “Data Analyst” Job position in “Bangalore” location.
# To scrape the job-title, job-location, company_name, experience_required.
# To scrape first 10 jobs data from the webpage https://www.naukri.com/
```

```python
# Step 1: First get the webpage https://www.naukri.com/

url_1 = 'https://www.naukri.com/'

driver.get(url_1)
```

```python
# Step 2 : Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the location” field.


search_job_1 = driver.find_element(By.CLASS_NAME,"suggestor-input ")

search_job_1.send_keys("Data Analyst")

search_locn_1 = driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")

search_locn_1.send_keys("Bangalore")                                
```

```python
# Step 3 :  Then click the search button.

search_btn_1 = driver.find_element(By.CLASS_NAME,"qsbSubmit")

search_btn_1.click()
```

```python
# Step 4:  Scrape the data for the first 10 jobs results.
# First 10 job titles
job_titles_1 = []
title_tags_1 = driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title_tags_1[0:10]:
    job_titles_1.append(i.text)
job_titles_1
```

```python
# First 10 job-location
location_1 = []
locn_tags_1 = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 locWdth"]')
for j in locn_tags_1[0:10]:
    location_1.append(j.text)
location_1
```

```python
# First 10 company_name
company_names_1 = []
comp_tags_1 = driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in comp_tags_1[0:10]:
    company_names_1.append(i.text)
company_names_1
```

```python
# First 10 experience_required
experience_1 = []
exp_tags_1 = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 expwdth"]')
for j in exp_tags_1[0:10]:
    experience_1.append(j.text)
experience_1
```

```python
# Checking the length of all lists
len(job_titles_1),len(location_1),len(company_names_1),len(experience_1)
```

```python
# Step 5:  Finally create a dataframe of the scraped data.
jobs_1 = pd.DataFrame(index=range(1,11))
jobs_1['Job Title'] = job_titles_1
jobs_1['Company '] = company_names_1
jobs_1['Location'] = location_1
jobs_1['Experience required'] = experience_1
jobs_1
```

```python
# Q7) Go to the link - https://www.myntra.com/shoes
# Set second Price filter and Color filter to “Black”
# And then scrape First 100 shoes data you get. The data should include “Brand” of the shoes , Short Shoe
# description, price of the shoe as shown in the below image.
```

```python
#  Go to the link - https://www.myntra.com/shoes

url_7 = "https://www.myntra.com/shoes"

driver.get(url_7)
```

```python
# Set Color filter to “Black”

apply_color_7 = driver.find_element(By.XPATH,'/html/body/div[2]/div/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')

apply_color_7.click()
```

```python
# Set second Price filter 

apply_price_7 = driver.find_element(By.XPATH,'/html/body/div[2]/div/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label')

apply_price_7.click()
```

```python
# Scraping the brand of the first 100 shoes

brand_7 = []
start = 0
end = 2
for page in range(start,end):
    brand_tag = driver.find_elements(By.XPATH,'//h3[@class="product-brand"]')
    for i in brand_tag:
        brand_7.append(i.text)
    next_btn_7 = driver.find_element(By.XPATH,'//li[@class="pagination-next"]')
    next_btn_7.click()
    time.sleep(2)
brand_7 = brand_7[0:100]
```

```python
# scrape next pages urls
page_urls = []

nxt_page = driver.find_elements(By.XPATH,"//ul[@class='pagination-container']/li/a")

for i in nxt_page:

    page_urls.append(i.get_attribute('href'))
```

```python
page_urls
```

```python
# Scraping shoe brand names from multiple pages

shoe_names = []
for url in page_urls[:3]:

    driver.get(url)

    Names=driver.find_elements(By.XPATH,"//div[@class='product-productMetaInfo']/h3")  #for scrapping shoe brand names

    for i in Names:

        shoe_names.append(i.text)
        
shoe_names = shoe_names[0:100]
```

```python
shoe_names
```

```python
# Scraping the short shoe description of the first 100 shoes

shoe_desc_7 = []
for url in page_urls[:3]:

    driver.get(url)

    desc=driver.find_elements(By.XPATH,'//h4[@class="product-product"]')  #for scrapping shoe description

    for i in desc:

        shoe_desc_7.append(i.text)

shoe_desc_7 = shoe_desc_7[0:100]
```

```python
shoe_desc_7
```

```python
# Scraping the price of the first 100 shoes

price_7 = []

for url in page_urls[:3]:

    driver.get(url)

    price=driver.find_elements(By.XPATH,'//div[@class="product-price"]')  #for scrapping shoe price

    for i in price:

        price_7.append(i.text)

    
price_7 = price_7[0:100]
```

```python
price_7
```

```python
#  Finally create a dataframe of the scraped data.
shoes_7 = pd.DataFrame(index=range(1,101))
shoes_7['Brand'] = shoe_names
shoes_7['Shoe Description '] = shoe_desc_7
shoes_7['Price'] = price_7

shoes_7
```
